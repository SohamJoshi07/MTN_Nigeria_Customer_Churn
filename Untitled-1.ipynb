```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
from sklearn.inspection import permutation_importance
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('mtn_customer_churn.csv')
df.head()   
```

```python
print('Data types:')
print(df.dtypes)

print('\nMissing values in each column:')
print(df.isnull().sum())

# Convert 'Date of Purchase' to datetime even though it is stored as a string
# This is important as date features often provide additional insights in time series analyses
if 'Date of Purchase' in df.columns:
    df['Date of Purchase'] = pd.to_datetime(df['Date of Purchase'], errors='coerce')

# Further cleaning steps could involve fixing inconsistent text (e.g., trim whitespace, lowercasing categorical text)
categorical_cols = ['State', 'MTN Device', 'Gender', 'Subscription Plan', 'Customer Churn Status']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()

# Optionally, preview cleaned data
df.head()
```

```python
numeric_df = df.select_dtypes(include=[np.number])

if not numeric_df.empty:
    # Histogram for each numeric variable
    numeric_cols = numeric_df.columns
    for col in numeric_cols:
        plt.figure(figsize=(6, 4))
        sns.histplot(numeric_df[col], kde=True, color='teal')
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()

    # If there are at least 4 numeric columns, produce a correlation heatmap
    if len(numeric_cols) >= 4:
        plt.figure(figsize=(10, 8))
        corr = numeric_df.corr()
        sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
        plt.title('Correlation Heatmap of Numeric Features')
        plt.show()
else:
    print('No numeric columns to visualize.')

if 'Customer Churn Status' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x='Customer Churn Status', palette='viridis')
    plt.title('Customer Churn Status Distribution')
    plt.xlabel('Churn Status')
    plt.ylabel('Count')
    plt.show()

# Pair plot on numeric features to visualize pairwise relationships (if enough numeric columns)
if len(numeric_df.columns) >= 2:
    sns.pairplot(numeric_df, diag_kind='kde', palette='husl')
    plt.show()
```

Chrun Predictor

```python
from sklearn.exceptions import NotFittedError

target_col = 'Customer Churn Status'
cols_to_drop = ['Customer ID', 'Date of Purchase','Full Name','Customer Reviews','Reason for Churn']

exsiting_drop_col = [col for col in cols_to_drop if col in df.columns]
predictor_col = [col for col in df.columns if col not in exsiting_drop_col]

if len(predictor_col)>0:
    df_model = pd.get_dummies(df[predictor_col], drop_first=True)
else:
    raise ValueError("No predictor columns available after dropping specified columns.")

X = df_model.copy()

if target_col in df.columns:
    y = df[target_col].apply(lambda x: 1 if x.strip().lower() == 'yes' else 0)
else:
    raise ValueError(f"Target column '{target_col}' not found in the dataframe.")

print('Feature matrix shape:', X.shape)
print('Target vector shape:', y.shape)

if X.shape[0] != y.shape[0]:
    raise ValueError("Number of samples in feature matrix and target vector do not match.")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print('Training set shape:', X_train.shape)
print('Test set shape:', X_test.shape)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

acuuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {acuuracy:.4f}')

## Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Churned', 'Churned'], yticklabels=['Not Churned', 'Churned'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()

## ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()

## Permutation Importance
perm_importance = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42)
sorted_idx = perm_importance.importances_mean.argsort()[::-1]
plt.figure(figsize=(8, 6))
plt.barh(X.columns[sorted_idx], perm_importance.importances_mean[sorted_idx], color='skyblue')
plt.xlabel('Mean Importance') 
plt.title('Permutation Feature Importance')
plt.show()

```
